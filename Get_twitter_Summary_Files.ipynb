# If you would like to merge the file into a single larger file

```python

import os
import pandas as pd 

path = os.path.join(wd, Summary_Files)  #This is the absolute path where files are located

Merged_files=pd.DataFrame()     #Create a empty data frame to store all the data
for root,dirs,files in os.walk(path):
  for file in files:
    if file.endswith(".csv"):      #only read .csv files
      data = pd.read_csv(file) 
      Merged_files = pd.concat([Merged_files, data])    #Merge by row the files  "data" into the "Merged_files"
Merged_files

```

# You can filter the  tweets using Pandas

```python


Just_English_Tweets= True #@param {type:"boolean"}
Just_with_Geolocation_information = False #@param {type:"boolean"}
Just_Original_tweets = False #@param {type:"boolean"}
Minimun_No_Likes = 3 #@param {type:"number"}
Minimun_No_Retweets =  0#@param {type:"number"}

Summary_Details2=Merged_files    # Creating a copy to avoid working on the orignal data frame
if(Just_English_Tweets):
  Summary_Details2=Summary_Details2[ Summary_Details2['Language']=='en']
if(Just_with_Geolocation_information):
  Summary_Details2=Summary_Details2[ Summary_Details2['Geolocation_coordinate']=='YES']
if(Just_Original_tweets):
  Summary_Details2=Summary_Details2[ Summary_Details2['RT']=='NO']


Summary_Details2=Summary_Details2[ Summary_Details2['Likes']>=Minimun_No_Likes]
Summary_Details2=Summary_Details2[ Summary_Details2['Retweets']>=Minimun_No_Retweets]

Summary_Details2

```
